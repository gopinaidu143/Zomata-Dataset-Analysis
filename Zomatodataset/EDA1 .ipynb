```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.size
```

```python
df_country.shape
```

```python
df_country.isnull().sum()
```

```python
[country for country in df_country.columns if df_country[country].isnull().sum()>0]
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
#top three countries uses zomato data
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

observation:zomato maximum records are from india(94%) them followed by UK and US with 1% and 4% 

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
#sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,hue='Rating color')
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['pink','red','orange','yellow','green','green'])
```

```python
#count plot to show frequency of colour
sns.countplot(x='Rating color',data=ratings,palette=['pink','red','orange','yellow','green','green'])
```
